```python
import pandas as pd
import datetime
```

```python
chapters = pd.read_csv('data/chapters.csv', parse_dates=['date']) 
characters = pd.read_csv('data/characters.csv') 
coc = pd.read_csv('data/coc.csv', keep_default_na=False) 
```

```python
chapters
```

```python
chapters['week'] = chapters.apply (lambda row: row['date'].isocalendar()[1], axis=1)
```

```python
chapters['year'] = chapters.apply (lambda row: row['date'].isocalendar()[0], axis=1)
```

```python
chapters
```

```python
year_week = {}
for year in range(min(chapters['year']), max(chapters['year']) + 1):
    year_week[year] = {}
    for week in range(min(chapters['week']), max(chapters['week']) + 1):
        year_week[year][week] = 0
year_week
```

```python
for index, chapter in chapters.iterrows(): 
#     print(chapter) 
    year_week[chapter['year']][chapter['week']] = 1
year_week
```
