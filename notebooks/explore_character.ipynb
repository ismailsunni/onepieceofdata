```python
import json
import pandas as pd
```

```python
path = '../data/characters_detail_parallel.json'
with open(path, 'r') as file:
    # Load the JSON data from the file
    data = json.load(file)
```

```python
print(data[0])
```

```python
new_data = []
for d in data:
    keys = d.keys()
    new_id = list(keys)[0]
    detail = d[new_id]
    detail['id'] = new_id
    new_data.append(detail)
print(new_data[0])
```

```python
count_null = 0
count_one = 0
count_other = 0
for d in new_data:
    if not d.get('origin'):
        count_null += 1
    elif len(d.get('origin')) == 1:
        count_one += 1
    else:
        count_other += 1
print(count_null, count_one, count_other)
```

```python
df = pd.DataFrame(new_data)
print(df)
```

```python
len(df.columns)
```

```python
city_counts = df['origin'].value_counts()
print(city_counts)
```

```python
def extract_single_element(lst):
    if not isinstance(lst, list):
        return lst
    if len(lst) == 0:
        return None
    return lst[0] if len(lst) == 1 else lst

dfx = df.applymap(extract_single_element)
```

```python
result_df = df[df['dfname'].notna()]
print(len(result_df))
```

```python
unique_values = dfx['origin'].unique()
print(unique_values)
```

```python
def contains_list(element):
    return isinstance(element, list)
contains_list_df = dfx.applymap(contains_list)
columns_with_lists = contains_list_df.any()
print(columns_with_lists[columns_with_lists].index.tolist())
```

```python
def get_region(origin):
    regions = ['East Blue', 'West Blue', 'North Blue', 'South Blue', 'Grand Line', 'Sky Islands']
    if not origin:
        return None
    if not isinstance(origin, str):
        return origin
    for r in regions:
        if r in origin:
            return r
    return origin

regions = dfx['origin'].apply(get_region)
# print(regions)

count_per_value = regions.value_counts().reset_index()

# Rename the columns for clarity
count_per_value.columns = ['result_value', 'count']
print(count_per_value)
```
